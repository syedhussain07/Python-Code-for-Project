```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Load the dataset
data = pd.read_csv('/content/startup_funding.csv')

# Data Cleaning
# Handling missing values for both numerical and categorical data
num_cols = data.select_dtypes(include=['float64', 'int64']).columns
cat_cols = data.select_dtypes(include=['object']).columns

data[num_cols] = data[num_cols].apply(lambda x: x.fillna(x.median()))
data[cat_cols] = data[cat_cols].apply(lambda x: x.fillna('Unknown'))

# Data Reduction with PCA for numerical columns
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[num_cols])

# Choosing a number of components for PCA (minimum of the number of numerical columns and 3)
n_components_pca = min(len(num_cols), 3)
pca = PCA(n_components=n_components_pca)
data_pca = pca.fit_transform(data_scaled)

# Creating a DataFrame for the PCA components
pca_cols = ['PCA' + str(i) for i in range(n_components_pca)]
data_pca_df = pd.DataFrame(data_pca, columns=pca_cols, index=data.index)

# Data Transformation: Normalizing numerical columns and encoding categorical columns
transformer = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(), cat_cols)
])

data_transformed = transformer.fit_transform(data)
transformed_cols = transformer.get_feature_names_out()

# Converting transformed data back to a DataFrame
data_transformed_df = pd.DataFrame(data_transformed.toarray(), columns=transformed_cols, index=data.index)

# Displaying the first five rows after each step
cleaned_data_head = data.head()
reduced_data_head = data_pca_df.head()
transformed_data_head = data_transformed_df.head()

cleaned_data_head, reduced_data_head, transformed_data_head
```
